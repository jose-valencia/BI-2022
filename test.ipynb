```python
import pandas as pd

data_portafolio = pd.read_csv("https://raw.githubusercontent.com/jose-valencia/BI-UCA/BI-2022/Lab-1/220517-Data/data-portafolio.csv",encoding = "Latin-1")
```

```python
r = data_portafolio
r.to_csv('drive/MyDrive/Colab Notebooks/original.csv')
r = data_portafolio.dropna(how='all')
r.to_csv('drive/MyDrive/Colab Notebooks/procesado.csv')
```

```python
import numpy as np
from matplotlib import pyplot as plt

y = np.random.randn(100)

x = [x for x in range(len(y))]

plt.plot(x, y, '-')
plt.fill_between(x, y, 200, where = (y > 195), facecolor='g', alpha=0.6)

plt.title("Sample Plot")
plt.show()
```

```python
x = np.asarray(data_portafolio['ID Casa'].dropna(how='all'))
y = np.asarray(data_portafolio['Precio oferta'].dropna(how='all'))

plt.plot(x, y, '-')

plt.title("Sample Plot")
plt.show()
```

```python
x = np.asarray(data_portafolio['ID Casa'].dropna(how='all'))
y = np.asarray(data_portafolio['Precio oferta'].dropna(how='all')) * 0.5
plt.plot(x, y, '-')

plt.title("Sample Plot")
plt.show()
```
